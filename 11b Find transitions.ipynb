```python
import json
import time
import sys
import sqlalchemy
from sqlalchemy import func
from sqlalchemy.orm import sessionmaker
```

```python
import database
from database.schema import Rating
```

```python
exp_name="find_transitions"
project_map_file = 'data/articles-projects.json'
output_file = "all_transitions.tsv"
project_output_file = "%d-%s.tsv"
```

```python
quality_names = [
    "FA-Class",
    "GA-Class",
    "A-Class",
    "Bplus-Class",
    "B-Class",
    "C-Class",
    "Stub-Class",
    "Start-Class",
    "Draft-Class",
    "Merge-Class",
    "Needed-Class",
    "NotA-Class",
    "Unassessed",
    "Unassessed-Class",
    "List-Class",
    "FL-Class",
    "AL-Class",
    "BL-Class",
    "CL-Class",
    "SL-Class",
    "FM-Class",
    "Category-Class",
    "Disambig-Class",
    "Portal-Class",
    "Template-Class",
    "Project-Class",
    "Book-Class",
    "File-Class",
    "Image-Class",
    "Redirect-Class",
    "Current-Class",
    "Future-Class",
    "NA-Class",
    ""
]
out_keys = ['old_quality',] + quality_names
```

```python
import logbook
exp = logbook.Experiment(exp_name)
log = exp.get_logger()
```

```python
log.info("Loading project json")
project_names = {}
with open(project_map_file, "rb") as f:
    for i, row in enumerate(f):
        datum = json.loads(row)
        project_names[datum["project_id"]] = datum["project_name"]
        if datum["project_id"] == 264:
            project_names[datum["project_id"]] = u'Wikipedia:WikiProject Pokémon'
log.info("Done loading project json")
```

```python
log.info("Connecting to database")
Session = sessionmaker()
Session.configure(bind=database.engine)
session = Session()
data = []
try:
    all_from = {}
    log.info("Calculating counts for all projects")
    for project_id, project_name in project_names.iteritems():
        log.info(
            "  Calculating counts for project: %d:%s" %
            (project_id, project_name))
        project_from = {}
        log.info("  Querying for rating changes")
        query = session.query(Rating.old_quality, Rating.new_quality) \
            .filter(Rating.project_id == project_id) \
            .filter(Rating.action.in_(["Assessed", "Reassessed"]))
        for (old_quality, new_quality) in query:
            if old_quality == new_quality:
                continue
            # Update project-wide counts
            if old_quality not in all_from:
                all_from[old_quality] = {}
            if new_quality not in all_from[old_quality]:
                all_from[old_quality][new_quality] = 0
            all_from[old_quality][new_quality] = all_from[old_quality][new_quality] + 1
            # Update per-project counts
            if old_quality not in project_from:
                project_from[old_quality] = {}
            if new_quality not in project_from[old_quality]:
                project_from[old_quality][new_quality] = 0
            project_from[old_quality][new_quality] = project_from[old_quality][new_quality] + 1
        # Write per-project stats
        name = project_names[project_id].replace('/', '.')
        filename = exp.get_filename(project_output_file % (project_id, name))
        with open(filename, "wb") as out:
            out.write(("\t".join(out_keys) + "\n").encode('utf-16-le'))
            for old_quality in quality_names:
                datum = [u"%s" % old_quality]
                for new_quality in quality_names:
                    try:
                        datum.append(u"%d" % project_from[old_quality][new_quality])
                    except KeyError:
                        datum.append(u"0")
                out.write(("\t".join([x for x in datum]) + "\n").encode('utf-16-le'))
except:
    log.error("Error: %s" % sys.exc_info()[0])
    raise
finally:
    session.close()
    log.info("Database session closed")
log.info("Writing transition data to output file")
filename = exp.get_filename(output_file)
with open(filename, "wb") as out:
    out.write(("\t".join(out_keys) + "\n").encode('utf-16-le'))
    for old_quality in quality_names:
        datum = [u"%s" % old_quality]
        for new_quality in quality_names:
            try:
                datum.append(u"%d" % all_from[old_quality][new_quality])
            except KeyError:
                datum.append(u"0")
        out.write(("\t".join([x for x in datum]) + "\n").encode('utf-16-le'))
```
