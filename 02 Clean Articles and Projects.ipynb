```python
import csv
import io
import json
import urllib
```

```python
infile = 'data/Article_Project_Names.csv'
projfile = 'data/projects.tsv'
metafile = 'data/meta_projects.tsv'
outfile = 'data/articles-projects.json'
articlefile = 'data/articles.json'
```

```python
exclude_namespaces = set([
    'User', 'User talk',
    'Wikipedia', 'Wikipedia talk',
    'WP', 'WP talk',
    'Project', 'Project talk',
    'File', 'File talk',
    'Image', 'Image talk',
    'MediaWiki', 'MediaWiki talk',
    'Template', 'Template talk',
    'Help', 'Help talk',
    'Category', 'Category talk',
    'Portal', 'Portal talk',
    'Book', 'Book talk',
    'Draft', 'Draft talk',
    'Education Program', 'Education Program talk',
    'TimedText', 'TimedText talk',
    'Module', 'Module talk',
    'Topic', 'Topic talk',
    'Gadget', 'Gadget talk',
    'Special', 'Special talk',
    'Media', 'Media talk'
])
proj_map = {}
with open(projfile, 'rb') as f:
    # Skip header
    f.next()
    for row in f:
        row = row[1:-1].decode('utf_16_le')
        if len(row) == 0:
            continue
        title, unique = row.split(u"\t")
        proj_map[title] = unique
meta = set()
with open(metafile, 'rU') as f:
    for line in f:
        m = line.rstrip("\n")
        meta.add(m)
        print m
out = {}
with open(infile,'rb') as f:
    reader = csv.reader(f)
    # Skip header
    reader.next()
    for i, row in enumerate(reader):
        if i % 100000 == 0:
            print "Article: %d" % (i)
        content = row[0].decode('utf-8')
        project = row[1].decode('utf-8')
        # Skip meta projects
        if project in meta:
            continue
        # Convert to unique project name
        project = proj_map[project]
        try:
            namespace, name = content.split(":", 1)
            if namespace in exclude_namespaces:
                continue
            else:
                name = "%s:%s" % (namespace, name)
        except ValueError:
            name = content
        if name[0:5] == "File:":
            print "content: %s" % content
            print content.split(":")
            namespace, name = content.split(":", 1)
            print namespace in exclude_namespaces
            raise KeyboardInterrupt
        try:
            out[project]['articles'].add(name)
        except KeyError:
            out[project] = {'project_name':project, 'articles':set()}
            out[project]['articles'].add(name)
with open(outfile, 'wb') as f:
    project_id = 1
    for project_name in out.iterkeys():
        out[project_name]["project_id"] = project_id
        project_id += 1
        out[project_name]["articles"] = list(out[project_name]["articles"])
        f.write(json.dumps(out[project_name]) + "\n")
```

```python
articles = {}
with open(outfile, 'rb') as f:
    for row in f:
        project = json.loads(row)
        for article in project['articles']:
            articles[article] = articles.get(article, [])
            articles[article].append(project['project_id'])
with open(articlefile, 'wb') as f:
    f.write(json.dumps(articles))
```
