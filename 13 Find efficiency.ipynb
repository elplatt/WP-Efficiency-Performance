```python
import json
import time
import sys
import sqlalchemy
from sqlalchemy import func
from sqlalchemy.orm import sessionmaker
```

```python
import database
from database.schema import Rating, revision_table
```

```python
exp_name="debug_13_find_efficiency"
project_map_file = 'data/articles-projects.json'
output_file = "efficiency.tsv"
```

```python
import logbook
exp = logbook.Experiment(exp_name)
log = exp.get_logger()
```

```python
log.info("Loading project json")
project_names = {}
with open(project_map_file, "rb") as f:
    for row in f:
        datum = json.loads(row)
        project_names[datum["project_id"]] = datum["project_name"]
log.info("Done loading project json")
```

```python
log.info("Connecting to database")
Session = sessionmaker()
Session.configure(bind=database.engine)
session = Session()
data = []
classes_a = set([
    "FA-Class",
    "GA-Class",
    "A-Class"
])
classes_b = set([
    "Bplus-Class",
    "B-Class"
])
classes_c = set([
    "C-Class"
])
classes_start = set([
    "",
    "Stub-Class",
    "Start-Class",
    "Draft-Class",
    "Merge-Class",
    "Needed-Class",
    "Unassessed",
    "Unassessed-Class",
    "Current-Class",
    "Future-Class"
])
classes_ab = classes_a | classes_b
classes_abc = classes_ab | classes_c
classes_other = set([
    "NotA-Class",
    "List-Class",
    "FL-Class",
    "AL-Class",
    "BL-Class",
    "CL-Class",
    "SL-Class",
    "FM-Class",
    "Category-Class",
    "Disambig-Class",
    "Portal-Class",
    "Template-Class",
    "Project-Class",
    "Book-Class",
    "File-Class",
    "Image-Class",
    "Redirect-Class",
    "NA-Class"
])
classes = classes_abc | classes_start | classes_other
try:
    log.info("Calculating efficiency for all projects")
    data = []
    #for project_id, project_name in project_names.iteritems():
    for project_id, project_name in [(1169, "Wikipedia:WikiProject National Football League")]:
        log.info(
            "  Calculating efficiency for project: %d:%s" %
            (project_id, project_name))
        ratings = {}
        rating_times = {}
        # The parameters below are defined in the paper
        Delta_a = 0
        Delta_b = 0
        Delta_c = 0
        N_a = 0
        N_b = 0
        N_c = 0
        log.info("  Opening table: %d_revisions" % project_id)
        revisions = revision_table(project_id)
        log.info("  Querying for rating changes")
        query = session.query(
                Rating.internal_id,
                Rating.action,
                Rating.article_name,
                Rating.old_quality,
                Rating.new_quality,
                Rating.new_article_name,
                Rating.timestamp) \
            .filter(Rating.project_id == project_id) \
            .filter(Rating.action.in_(["Assessed", "Reassessed", "Renamed", "Removed"])) \
            .order_by(Rating.timestamp)
        for row in query:
            action = row[1]
            article_name = row[2]
            timestamp = row[6]
            if action in ["Assessed", "Reassessed"]:
                # Could be rating or importance
                old_rating = row[3]
                new_rating = row[4]
                if new_rating == old_rating:
                    # No grade transition, keep going until we find one
                    continue
                try:
                    old_rating = ratings[article_name]["rating"]
                except KeyError:
                    old_rating = ''
                a_transition = 0
                b_transition = 0
                c_transition = 0
                if new_rating in classes_a and old_rating not in classes_a:
                    a_transition = 1
                elif old_rating in classes_a and new_rating not in classes_a:
                    a_transition = -1
                if new_rating in classes_ab and old_rating not in classes_ab:
                    b_transition = 1
                elif old_rating in classes_ab and new_rating not in classes_ab:
                    b_transition = -1
                if new_rating in classes_abc and old_rating not in classes_abc:
                    c_transition = 1
                elif old_rating in classes_abc and new_rating not in classes_abc:
                    c_transition = -1
                if a_transition == 0 and b_transition == 0 and c_transition == 0:
                    try:
                        ratings[article_name]["rating"] = new_rating
                    except KeyError:
                            ratings[article_name] = {
                                "rating": new_rating, "timestamp": timestamp }
                    continue
                # There has been a transition, record it and count the revision
                Delta_a += a_transition
                Delta_b += b_transition
                Delta_c += c_transition
                
                revision_q = session.query(func.count()) \
                    .filter(revisions.article_name.in_([article_name, "Talk:%s" % article_name]))
                # Filter by time period
                # Revisions on assessment days are associated with the old rating
                revision_q = revision_q.filter(revisions.timestamp <= timestamp)
                try:
                    old_timestamp = ratings[article_name]["timestamp"]
                    revision_q = revision_q.filter(revisions.timestamp > old_timestamp)
                except KeyError:
                    pass
                count = revision_q.first()[0]
                log.info("%s %s->%s: %d edits" % (article_name, old_rating, new_rating, count))
                # N_g counts the number of revisions that occurred belowe rating `g`
                if old_rating in classes_b:
                    N_a += count
                elif old_rating in classes_c:
                    N_a += count
                    N_b += count
                elif old_rating in classes_start:
                    N_a += count
                    N_b += count
                    N_c += count
                ratings[article_name] = {
                    "rating": new_rating, "timestamp": timestamp }
            elif row[1] == "Renamed":
                new_article_name = row[4]
                revision_q = session.query(func.count()) \
                    .filter(revisions.article_name.in_([article_name, "Talk:%s" % article_name]))
                # Filter by time period
                # Revisions on assessment days are associated with the old rating
                revision_q = revision_q.filter(revisions.timestamp <= timestamp)
                try:
                    old_timestamp = ratings[article_name]["timestamp"]
                    revision_q = revision_q.filter(revisions.timestamp > old_timestamp)
                except KeyError:
                    pass
                count = revision_q.first()[0]
                # N_g counts the number of revisions that occurred belowe rating `g`
                if old_rating in classes_b:
                    N_a += count
                elif old_rating in classes_c:
                    N_a += count
                    N_b += count
                elif old_rating in classes_start:
                    N_a += count
                    N_b += count
                    N_c += count
                # Update timestamp and rename article
                ratings[new_article_name] = {
                    "rating": old_rating, "timestamp": timestamp }
                try:
                    del ratings[article_name]
                except KeyError:
                    log.error("    Renamed unknown article: %d: %s" %
                        (row[0], article_name))
            elif row[1] == "Removed":
                try:
                    del ratings[article_name]
                except KeyError:
                    log.error("    Removed unknown article: %d: %s" %
                        (row[0], article_name))
        datum = [project_id, Delta_a, Delta_b, Delta_c, N_a, N_b, N_c]
        data.append(datum)
except:
    log.error("Error: %s" % sys.exc_info()[0])
    raise
finally:
    session.close()
    log.info("Database session closed")
log.info("Writing efficiency data to output file")
with open(exp.get_filename(output_file), "wb") as out:
    out.write("\t".join(
        ['project_id', 'Delta_a', 'Delta_b', 'Delta_c', 'N_a', 'N_b', 'N_c']) + "\n")
    for row in data:
        out.write(",".join([str(x) for x in row]) + "\n")
log.info("Done.")
```
